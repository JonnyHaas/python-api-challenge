### Step 1: Create a map that displays a point for every city in the `city_data_df` DataFrame. The size of the point should be the humidity in each city.

```python
import folium
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Function to convert matplotlib color to hex
def rgb_to_hex(rgb_color):
    return mcolors.to_hex(rgb_color)

# Create a colormap
cmap = plt.cm.get_cmap('rainbow')
norm = mcolors.Normalize(vmin=city_data_df['Humidity'].min(), vmax=city_data_df['Humidity'].max())

# Create a map centered around the mean latitude and longitude
map_center = [city_data_df['Lat'].mean(), city_data_df['Lng'].mean()]
mymap = folium.Map(location=map_center, zoom_start=4, max_bounds=True)

# Add points to the map
for _, row in city_data_df.iterrows():
    color = rgb_to_hex(cmap(norm(row['Humidity'])))
    folium.CircleMarker(
        location=[row['Lat'], row['Lng']],
        radius=max(2, row['Humidity'] * 0.05),  # Adjust size relative to humidity, with a minimum size
        color=color,  # Use the rainbow colormap for color
        fill=True,
        fill_color=color,
        fill_opacity=0.7
    ).add_to(mymap)

# Set the map's min and max zoom levels
mymap.options.update({
    "minZoom": 2,
    "maxZoom": 16
})

# Display the map
mymap
```

### Step 2: Narrow down the `city_data_df` DataFrame to find your ideal weather condition

```python
# Assuming city_data_df is already loaded

# Apply the filters to the DataFrame
ideal_conditions_df = city_data_df[
    (city_data_df['Max Temp'] > 21) & 
    (city_data_df['Max Temp'] < 27) & 
    (city_data_df['Wind Speed'] < 4.5) & 
    (city_data_df['Cloudiness'] == 0)
]

# Display the DataFrame with ideal weather conditions
ideal_conditions_df
```

```python
import folium
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Function to convert matplotlib color to hex
def rgb_to_hex(rgb_color):
    return mcolors.to_hex(rgb_color)

# Create a colormap
cmap = plt.cm.get_cmap('rainbow')
norm = mcolors.Normalize(vmin=ideal_conditions_df['Humidity'].min(), vmax=ideal_conditions_df['Humidity'].max())

# Create a map centered around the mean latitude and longitude of the ideal conditions
map_center = [ideal_conditions_df['Lat'].mean(), ideal_conditions_df['Lng'].mean()]
mymap = folium.Map(location=map_center, zoom_start=4, max_bounds=True)

# Add points to the map for cities with ideal conditions
for _, row in ideal_conditions_df.iterrows():
    color = rgb_to_hex(cmap(norm(row['Humidity'])))
    folium.CircleMarker(
        location=[row['Lat'], row['Lng']],
        radius=max(2, row['Humidity'] * 0.05),  # Adjust size relative to humidity, with a minimum size
        color=color,  # Use the rainbow colormap for color
        fill=True,
        fill_color=color,
        fill_opacity=0.7
    ).add_to(mymap)

# Set the map's min and max zoom levels
mymap.options.update({
    "minZoom": 2,
    "maxZoom": 16
})

# Display the map
mymap
```

### Step 3: Create a new DataFrame called `hotel_df`.

```python

# Select the necessary columns and create a new DataFrame called hotel_df
hotel_df = ideal_conditions_df[['City', 'Country', 'Lat', 'Lng', 'Humidity']].copy()

# Add an empty 'Hotel Name' column to hotel_df
hotel_df['Hotel Name'] = ""

# Display the DataFrame with an empty 'Hotel Name' column
hotel_df
```

### Step 4: For each city, use the Geoapify API to find the first hotel located within 10,000 metres of your coordinates.

```python
import folium
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

# Function to convert matplotlib color to hex
def rgb_to_hex(rgb_color):
    return mcolors.to_hex(rgb_color)

# Create a colormap
cmap = plt.cm.get_cmap('rainbow')
norm = mcolors.Normalize(vmin=hotel_df['Humidity'].min(), vmax=hotel_df['Humidity'].max())

# Create a map centered around the mean latitude and longitude
map_center = [hotel_df['Lat'].mean(), hotel_df['Lng'].mean()]
mymap = folium.Map(location=map_center, zoom_start=4, max_bounds=True, no_wrap=True)

# Add hotel markers to the map with colored and sized points
for _, row in hotel_df.iterrows():
    color = rgb_to_hex(cmap(norm(row['Humidity'])))
    folium.CircleMarker(
        location=[row['Lat'], row['Lng']],
        radius=max(2, row['Humidity'] * 0.05),  # Adjust size relative to humidity, with a minimum size
        popup=f"{row['Hotel Name']} - {row['City']}, {row['Country']}",
        color=color,
        fill=True,
        fill_color=color,
        fill_opacity=0.7
    ).add_to(mymap)

# Set the map's min and max zoom levels
mymap.options.update({
    "minZoom": 2,
    "maxZoom": 16
})

# Display the map
mymap


```

### Step 5: Add the hotel name and the country as additional information in the hover message for each city in the map.

```python
import pandas as pd
import plotly.express as px

# Assuming hotel_df is already defined and contains the required columns

# Use Plotly Express to create the map
fig = px.scatter_geo(hotel_df,
                     lat='Lat',
                     lon='Lng',
                     color='Humidity',
                     hover_name='City',
                     hover_data={'Country': True, 'Hotel Name': True, 'Humidity': True, 'Lat': False, 'Lng': False},
                     color_continuous_scale=px.colors.sequential.Rainbow,  # Rainbow color scale
                     title='Hotel Locations with Ideal Weather Conditions')

# Update the layout for a single rotation of Earth and style similar to previous script
fig.update_geos(
    projection_rotation=dict(lon=180),
    showcoastlines=True,
    coastlinecolor="DarkBlue",
    showland=True,
    landcolor="LightGreen",
    showocean=True,
    oceancolor="LightBlue",
    projection_type="natural earth",
    lataxis_showgrid=True,
    lonaxis_showgrid=True
)

# Update layout configurations to match previous script style
fig.update_layout(
    margin=dict(l=0, r=0, t=50, b=0),
    geo=dict(
        bgcolor='rgba(0,0,0,0)',  # Transparent background
        lakecolor='LightBlue',  # Light blue lakes
        landcolor='LightGreen',  # Light green land
    ),
    coloraxis_colorbar=dict(
        title='Humidity',
        tickvals=hotel_df['Humidity'],
        ticktext=hotel_df['Humidity']
    )
)

# Display the figure
fig.show()
```
